```python
# is used to check for missing values (NaN) in each column of the DataFrame
df.isnull().sum()
```

```python
# is used to check for any duplicates in the rows
df.duplicated().sum()
```

```python
# removes any duplicate rows from the DataFrame
df.drop_duplicates(inplace=True)
```

```python
max_id = df["OrderID"].max()
print("The maximum value in OrderID:",max_id)

missing_indices = df[df["OrderID"].isna()].index
print("indexes of NaN value:",missing_indices)

# Fill missing values with max_id + 1, +2, ...
for i, idx in enumerate(missing_indices):
    df.at[idx, "OrderID"] = max_id + i + 1

```

```python
# Fill with a placeholder (-1) and convert to integer.
df['OrderID'] = df['OrderID'].fillna(-1).astype(int)

# 2. Fill UnitPrice missing values with median price per Product
df["UnitPrice"] = df.groupby("Product")["UnitPrice"].transform(lambda x: x.fillna(x.median()))


# 3. Fill SalesPersonID missing values with the most frequent ID (mode)
df["SalesPersonID"] = df["SalesPersonID"].fillna(df["SalesPersonID"].mode()[0])
```

```python
# After the update the Postal Code it will be 0
df.isnull().sum()
```

```python
# It shows a list of the columns names
print(df.columns)
```

**top 5 **

```python
# It Top 5 Products by Number of Sales
top_products = df.groupby('Product')['OrderQty'].sum().reset_index()
top_5_products = top_products.sort_values(by='OrderQty', ascending=False).head(5)

print("Top 5 Products by Number of Sales:")
print(top_5_products.to_markdown())

plot_data = top_5_products.sort_values(by='OrderQty', ascending=True)

plt.barh(plot_data['Product'], plot_data['OrderQty'], color='skyblue')
plt.xlabel('Total Order Quantity (Number of Sales)')
plt.ylabel('Product')
plt.title('Top 5 Products by Number of Sales')
```

**bottom**

```python
# It shows Lowest 5 Products by Number of Sales
product_sales_qty = df.groupby('Product')['OrderQty'].sum().reset_index()
product_sales_qty.columns = ['Product Name', 'Order Quantity']

lowest_5_products = product_sales_qty.sort_values(by='Order Quantity', ascending=True).head(5)

print("Lowest 5 Products by Number of Sales:")
print(lowest_5_products.to_markdown())

plot_data = lowest_5_products.sort_values(by='Order Quantity')

plt.barh(plot_data['Product Name'], plot_data['Order Quantity'], color='orange')
plt.xlabel('Total Order Quantity (Number of Sales)')
plt.ylabel('Product Name')
plt.title('Lowest 5 Products by Number of Sales')
```

**category**

```python
# finds the highest and lowest sale by grouping the category values (Accessories,Bikes , etc..) and then you can see the max and min

top_in_category = df.groupby("ProductCategory")["LineTotal"].max()
bottom_in_category = df.groupby("ProductCategory")["LineTotal"].min()

top_in_category.index.name = "Category"
top_in_category.name = "Max Sales (LineTotal)"
bottom_in_category.index.name = "Category"
bottom_in_category.name = "Min Sales (LineTotal)"

print("\nTop in each Category:\n", top_in_category.to_markdown())
print("\nBottom in each Category:\n", bottom_in_category.to_markdown())
```
